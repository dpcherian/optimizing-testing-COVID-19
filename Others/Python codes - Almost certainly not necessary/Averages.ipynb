```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
begin_at_0_purePCR = glob.glob('../data/TargettedTesting/StartingAt0-Locking/OnlyI-Targetted_Testing_BeginAt_0_DTR_0.1_RAT_0.75_0.0_PCR_1.0_1.0_*')
begin_at_0_pureRAT = glob.glob('../data/TargettedTesting/StartingAt0-Locking/OnlyI-Targetted_Testing_BeginAt_0_DTR_0.1_RAT_0.75_1.0_PCR_1.0_0.0_*')


begin_at_1_purePCR = glob.glob('../data/TargettedTesting/OnlyI-Targetted_Testing_BeginAt_1_DTR_0.1_RAT_0.75_0.0_PCR_1.0_1.0_*')
begin_at_1_pureRAT = glob.glob('../data/TargettedTesting/OnlyI-Targetted_Testing_BeginAt_1_DTR_0.1_RAT_0.75_1.0_PCR_1.0_0.0_*')
begin_at_1_equalmix= glob.glob('../data/TargettedTesting/OnlyI-Targetted_Testing_BeginAt_1_DTR_0.1_RAT_0.75_0.5_PCR_1.0_0.5_*')

begin_at_10_purePCR = glob.glob('../data/TargettedTesting/OnlyI-Targetted_Testing_BeginAt_10_DTR_0.1_RAT_0.75_0.0_PCR_1.0_1.0_*')
begin_at_10_pureRAT = glob.glob('../data/TargettedTesting/OnlyI-Targetted_Testing_BeginAt_10_DTR_0.1_RAT_0.75_1.0_PCR_1.0_0.0_*')
begin_at_10_equalmix= glob.glob('../data/TargettedTesting/OnlyI-Targetted_Testing_BeginAt_10_DTR_0.1_RAT_0.75_0.5_PCR_1.0_0.5_*')
```

```python
print(len(begin_at_0_purePCR))
print(len(begin_at_0_pureRAT))
print(len(begin_at_1_purePCR))
print(len(begin_at_1_pureRAT))
print(len(begin_at_1_equalmix))

print(len(begin_at_10_purePCR))
print(len(begin_at_10_pureRAT))
print(len(begin_at_10_equalmix))
```

```python
avg_0_pcr = np.zeros((121,9))

for i in range(len(begin_at_0_purePCR)):
    file = np.genfromtxt(begin_at_0_purePCR[i],delimiter=" ", comments="#")
    avg_0_pcr += file
    plt.plot(file[:,0], file[:,2]+file[:,3]+file[:,5],color='firebrick',alpha=0.2)
    
avg_0_pcr = avg_0_pcr/len(begin_at_0_purePCR)
plt.plot(avg_0_pcr[:,0], avg_0_pcr[:,2]+avg_0_pcr[:,3]+avg_0_pcr[:,5],color='firebrick')

np.savetxt('./averageRuns_StartTestingAt_0_DTR_0.1_PurePCR',avg_0_pcr,fmt="%i")
plt.ylabel('Number of people')
plt.xlabel('Time (Days)')
plt.title("StartTestingAt_0_DTR_0.1_PurePCR")
plt.show()

avg_0_rat = np.zeros((121,9))

for i in range(len(begin_at_0_pureRAT)):
    file = np.genfromtxt(begin_at_0_pureRAT[i],delimiter=" ", comments="#")
    avg_0_rat += file
    plt.plot(file[:,0], file[:,2]+file[:,3]+file[:,5],color='firebrick',alpha=0.2)
    
avg_0_rat = avg_0_rat/len(begin_at_0_pureRAT)
plt.plot(avg_0_rat[:,0], avg_0_rat[:,2]+avg_0_rat[:,3]+avg_0_rat[:,5],color='firebrick')
np.savetxt('./averageRuns_StartTestingAt_0_DTR_0.1_PureRAT',avg_0_rat,fmt="%i")
plt.ylabel('Number of people')
plt.xlabel('Time (Days)')
plt.title("StartTestingAt_0_DTR_0.1_PureRAT")
plt.show()
```
