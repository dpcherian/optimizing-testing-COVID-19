```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Cdata = pd.read_csv("./montecarlo_SAIRH_highdensity_test_10_99_1hr.txt",delimiter=" ", header=None)
# PGBdata = pd.read_csv("./BACKUP_GA_SAIRH.txt",delimiter=" ", header=None)
PGNdata = pd.read_csv("./python-pop1000.txt",delimiter=" ", header=None)

# PMdata = pd.read_csv("./tau_mc_SAIRH_d.txt",delimiter=" ", header=None)
```

```python
labels = ["t","Susceptible","Asymptomatic","Infected","Recovered","Hospitalised"]

for i in range(1,len(labels)):
    plt.plot(Cdata[0],Cdata[i], label= labels[i][0]+" "+" C++")
#     plt.plot(PGBdata[0],PGBdata[i], label= labels[i][0]+" "+" GABPyn")
    plt.plot(PGNdata[0],PGNdata[i], label= labels[i][0]+" "+" GANPyn")
#     plt.plot(PMdata[0],PMdata[i], label= labels[i][0]+" "+" MCPyn")
    plt.xlabel("Time (Days)")
    plt.ylabel("No. of "+labels[i])
    plt.xlim(0,100)
    plt.legend()
    plt.show()
```
