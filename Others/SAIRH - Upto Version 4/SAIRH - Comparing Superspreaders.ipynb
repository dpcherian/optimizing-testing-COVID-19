```python
import numpy as np
import matplotlib.pyplot as plt


import pandas as pd
import glob

import matplotlib.patches as mpatches
```

```python
txt_files = np.array(glob.glob('../data/Superspreader_1/*.txt'))
```

```python
print(txt_files)
```

```python
endpoints_superspreader = np.zeros((5,4),int)
```

```python
pcr = []

for i in range(0,len(txt_files)):
    if(txt_files[i][24:27]=="PCR"):
        pcr.append(txt_files[i])

pcr_none = []
pcr_5 = []
pcr_10 = []
pcr_50 = []

all_pcr_data = [pcr_none,pcr_5,pcr_10,pcr_50]

for i in range(0,len(pcr)):
    if(pcr[i][28:31] == "0.0"):
        pcr_none.append(pcr[i])
    elif(pcr[i][28:31] == "5.0"):
        pcr_5.append(pcr[i])
    elif(pcr[i][28:31] == "10."):
        pcr_10.append(pcr[i])
    elif(pcr[i][28:31] == "50."):
        pcr_50.append(pcr[i])

clrs = ["maroon", "red", "darkgoldenrod", "darkorange"] 
pcr_labels = ["PCR - No testing", "PCR - 5% testing", "PCR - 10% testing", "PCR - 50% testing"]

pcr_avgs = []


plt.figure(figsize=(10,10))
plt.title("Superspreader - PCR - 100% Sensitivity",fontsize=20)
for i in range(0,len(all_pcr_data)):
    sum_val = 0
    for j in range(0,len(all_pcr_data[i])):
        temp_file = np.genfromtxt(all_pcr_data[i][j],delimiter=" ")
        #plt.plot(temp_file[:,0],temp_file[:,2]+temp_file[:,3]+temp_file[:,5],color=clrs[i],alpha=0.3)
        plt.plot(temp_file[:,0],temp_file[:,4],color=clrs[i],alpha=0.4,ls='--')
        
        sum_val += temp_file[:,4]
    
    if(len(all_pcr_data[i])!=0):
        sum_val = sum_val/len(all_pcr_data[i])
        endpoints_superspreader[0][i] = sum_val[-1]
        plt.plot(temp_file[:,0],sum_val,color=clrs[i],label=pcr_labels[i])
    
plt.xlabel("Time (Days)",fontsize=20)
plt.ylabel("Number of recovered",fontsize=20)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
#plt.yscale('log')
#plt.xlim(-1,200)
plt.savefig("../images/Superspreader/PCR.png",facecolor='w')
plt.show()
```
