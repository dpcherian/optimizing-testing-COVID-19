```python
import numpy as np
import matplotlib.pyplot as plt


import pandas as pd
import glob

import matplotlib.patches as mpatches
```

```python
txt_files = glob.glob('../data/*.txt')
```

```python
pcr_files_pointone = []
pcr_files_one      = []
pcr_files_five     = []
pcr_files_seven    = []
pcr_files_ten      = []

rat_files_pointone = []
rat_files_one      = []
rat_files_five     = []
rat_files_seven    = []
rat_files_ten      = []

for i in range(0,len(txt_files)):
    if(txt_files[i][8:11]=="PCR"):
        if(txt_files[i][12:15]=="0.1"):
            pcr_files_pointone.append(txt_files[i])
        elif(txt_files[i][12:15]=="1.0"):
            pcr_files_one.append(txt_files[i])
        elif(txt_files[i][12:15]=="5.0"):
            pcr_files_five.append(txt_files[i])
        elif(txt_files[i][12:15]=="7.5"):
            pcr_files_seven.append(txt_files[i])
        elif(txt_files[i][12:15]=="10."):
            pcr_files_ten.append(txt_files[i])
    elif(txt_files[i][8:11]=="RAT"):
        if(txt_files[i][12:15]=="0.1"):
            rat_files_pointone.append(txt_files[i])
        elif(txt_files[i][12:15]=="1.0"):
            rat_files_one.append(txt_files[i])
        elif(txt_files[i][12:15]=="5.0"):
            rat_files_five.append(txt_files[i])
        elif(txt_files[i][12:15]=="7.5"):
            rat_files_seven.append(txt_files[i])
        elif(txt_files[i][12:15]=="10."):
            rat_files_ten.append(txt_files[i])
            
pcr_files_pointone = sorted(pcr_files_pointone)
pcr_files_one      = sorted(pcr_files_one)
pcr_files_five     = sorted(pcr_files_five)
pcr_files_seven    = sorted(pcr_files_seven)
pcr_files_ten      = sorted(pcr_files_ten)

rat_files_pointone = sorted(rat_files_pointone)
rat_files_one      = sorted(rat_files_one)
rat_files_five     = sorted(rat_files_five)
rat_files_seven    = sorted(rat_files_seven)
rat_files_ten      = sorted(rat_files_ten)

# print(pcr_files_pointone)
# print(pcr_files_one)
# print(pcr_files_ten)

all_pcr_files = [pcr_files_pointone,pcr_files_one,pcr_files_five,pcr_files_seven,pcr_files_ten]
all_rat_files = [rat_files_pointone,rat_files_one,rat_files_five,rat_files_seven,rat_files_ten]
```

```python
pcr_pointone = []
pcr_one      = []
pcr_five     = []
pcr_seven    = []
pcr_ten      = []

rat_pointone = []
rat_one      = []
rat_five     = []
rat_seven    = []
rat_ten      = []


all_pcr_data = [pcr_pointone,pcr_one,pcr_five,pcr_seven,pcr_ten]
all_rat_data = [rat_pointone,rat_one,rat_five,rat_seven,rat_ten]

all_pcr_type = [[],[],[],[],[]]
all_rat_type = [[],[],[],[],[]]

n_pop = 10000


for i in range(0,len(all_pcr_files)):
    for j in range(0,len(all_pcr_files[i])):
        temp = pd.read_csv(all_pcr_files[i][j],delimiter=" ", header=None)[4][1001]/n_pop
        all_pcr_data[i].append(temp)
        
        temp_type = all_pcr_files[i][j][16:22]
        all_pcr_type[i].append(temp_type)
        
for i in range(0,len(all_rat_files)):
    for j in range(0,len(all_rat_files[i])):
        temp = pd.read_csv(all_rat_files[i][j],delimiter=" ", header=None)[4][1001]/n_pop
        all_rat_data[i].append(temp)
        
        temp_type = all_rat_files[i][j][16:22]
        all_rat_type[i].append(temp_type)


        

x_axis_pcr = [[0.1]*len(pcr_pointone),[1.0]*len(pcr_one),[5.0]*len(pcr_five),[7.5]*len(pcr_seven),[10.0]*len(pcr_ten)]
x_axis_rat = [[0.1]*len(rat_pointone),[1.0]*len(rat_one),[5.0]*len(rat_five),[7.5]*len(rat_seven),[10.0]*len(rat_ten)]

labels = ["0.1","1.0","5.0","7.5","10.0"]

pcr_colorscheme = np.array([["mediumvioletred","palevioletred"],     # 0.1 (Random testing, Symptomatic testing)
                   ["dodgerblue","skyblue"],                # 1.0 (Random testing, Symptomatic testing)
                   ["grey","red"],                          # 5.0(Random testing, Symptomatic testing)
                   ["green", "black"],                      # 7.5
                   ["yellow", "cyan"]])                     # 10.0   


pcr_colorscheme = np.array([["grey","red"],     # 0.1 (Random testing, Symptomatic testing)
                   ["grey","red"],                # 1.0 (Random testing, Symptomatic testing)
                   ["grey","red"],                          # 5.0(Random testing, Symptomatic testing)
                   ["grey","red"],                      # 7.5
                   ["grey","red"]])                     # 10.0   




rat_colorscheme = np.array([["orangered","tomato"],     # 0.1 (Random testing, Symptomatic testing)
                   ["navy","blue"],                # 1.0 (Random testing, Symptomatic testing)
                   ["grey","red"],                        # 5.0(Random testing, Symptomatic testing)
                   ["green", "black"],                      # 7.5
                   ["yellow", "cyan"]])                     # 10.0

rat_colorscheme = np.array([["grey","red"],     # 0.1 (Random testing, Symptomatic testing)
                   ["grey","red"],                # 1.0 (Random testing, Symptomatic testing)
                   ["grey","red"],                          # 5.0(Random testing, Symptomatic testing)
                   ["grey","red"],                      # 7.5
                   ["grey","red"]])                     # 10.0   





plt.figure(figsize=(20,10))
plt.title("PCR test results",fontsize='20')

for i in range(0, len(all_pcr_data)):
    for j in range(0, len(all_pcr_data[i])):
        if(all_pcr_type[i][j] == "Random"):
            col = pcr_colorscheme[:,0][i]
            m = 'o'
        elif(all_pcr_type[i][j] == "Sympto"):
            col = pcr_colorscheme[:,1][i]
            m = 'o'
        plt.scatter(x_axis_pcr[i][j],all_pcr_data[i][j],color=col,s=100,marker=m)#,label="PCR "+all_pcr_type[i][j]+" "+labels[i]

# plt.legend(fontsize='15')
plt.xlabel("Fraction of population tested daily",fontsize=20)
plt.ylabel("Asymptotic Fraction Infected",fontsize=20)
plt.locator_params(nbins = len(labels),axis='x')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

grey_patch = mpatches.Patch(color='grey', label='Random Testing')
red_patch = mpatches.Patch(color='red', label='Targeted Symptomatic Testing')
plt.legend(handles=[grey_patch,red_patch], fontsize=20)



plt.savefig("../images/PCR_comparison_1.png",facecolor='w')
plt.show()

plt.figure(figsize=(20,10))
plt.title("RAT test results",fontsize='20')


for i in range(0, len(all_rat_data)):
    for j in range(0, len(all_rat_data[i])):
        if(all_rat_type[i][j] == "Random"):
            col = rat_colorscheme[:,0][i]
            m='x'
        elif(all_rat_type[i][j] == "Sympto"):    
            col = rat_colorscheme[:,1][i]
            m='x'
        plt.scatter(x_axis_rat[i][j],all_rat_data[i][j],color=col,s=100,marker=m)#,label="RAT "+all_pcr_type[i][j]+" "+labels[i])
        
#plt.legend(fontsize='15')
plt.xlabel("Fraction of population tested daily",fontsize=20)
plt.ylabel("Asymptotic Fraction Infected",fontsize=20)
plt.locator_params(nbins = len(labels),axis='x')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

grey_patch = mpatches.Patch(color='grey', label='Random Testing')
red_patch = mpatches.Patch(color='red', label='Targeted Symptomatic Testing')
plt.legend(handles=[grey_patch,red_patch], fontsize=20)

plt.savefig("../images/RAT_comparison_1.png",facecolor='w')
plt.show()
```

```python
pcr_random = [[],[],[],[],[]]
pcr_symp = [[],[],[],[],[]]

for i in range(0, len(all_pcr_data)):
    
    for j in range(0, len(all_pcr_data[i])):
        
        if(all_pcr_type[i][j] == "Random"):
            pcr_random[i].append(all_pcr_data[i][j])
            
        elif(all_pcr_type[i][j] == "Sympto"):
            pcr_symp[i].append(all_pcr_data[i][j])


rat_random = [[],[],[],[],[]]
rat_symp = [[],[],[],[],[]]
            
for i in range(0, len(all_rat_data)):
    
    for j in range(0, len(all_rat_data[i])):
        
        if(all_rat_type[i][j] == "Random"):
            rat_random[i].append(all_rat_data[i][j])
            
        elif(all_rat_type[i][j] == "Sympto"):
            rat_symp[i].append(all_rat_data[i][j])

pcr_random = list(filter(None, pcr_random))
pcr_symp = list(filter(None, pcr_symp))

rat_random = list(filter(None, rat_random))
rat_symp = list(filter(None, rat_symp))


pcr_random_avg = np.average(np.array(pcr_random,dtype=object),axis=1)
pcr_symp_avg = np.average(np.array(pcr_symp,dtype=object),axis=1)

rat_random_avg = np.average(np.array(rat_random,dtype=object),axis=1)
rat_symp_avg = np.average(np.array(rat_symp,dtype=object),axis=1)

pcr_random_std = np.std(np.array(pcr_random),axis=1)
pcr_symp_std = np.std(np.array(pcr_symp),axis=1)

rat_random_std = np.std(np.array(rat_random),axis=1)
rat_symp_std = np.std(np.array(rat_symp),axis=1)

print(pcr_random_std)
```
