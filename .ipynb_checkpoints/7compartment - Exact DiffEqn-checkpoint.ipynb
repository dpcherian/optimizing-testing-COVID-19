```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
rate = np.zeros((7,7))
s = 0
a = 1
p = 2
mi= 3
si= 4
r = 5
h = 6

n0 = np.array([9990,10,0,0,0,0,0])

k_S   = 0.25;
gamma = 0.5;  # Fraction going from S->A

k_A   = 0.143;
k_P   = 0.5;
delta = 0.85;  # Fraction going from P->MI

k_MI  = 0.1;
k_SI  = 0.5;
sigma = 0.8;

k_H = 0.1;

rate[s][a] = gamma*k_S;        # S -> A
rate[s][p] = (1 - gamma)*k_S;  # S -> I
rate[a][r] = k_A;              # A -> R
rate[p][mi]= delta*k_P;        # P -> MI
rate[p][si]= (1-delta)*k_P;    # P -> SI
rate[mi][r]= k_MI;             # MI -> R
rate[si][r]= (1-sigma)*k_SI;   # SI -> R
rate[si][h]= sigma*k_SI;       # SI -> H
rate[h][r] = k_H;              # H -> R

Cpars = np.array([1,  1,   1,    1,   1,  0.1])
                # SA  SP  S-MI  S-SI  S-H   S-Q
    
```

```python
def an_soln(rate,Cpars,n0,T_max):
    
    n = np.copy(n0)
     
    N = np.sum(n0)
    print(N)
    t = np.linspace(0, T_max, T_max)# A grid of time points (in days)
    
    #########################################

    # The differential equations ############
    
    def deriv(y, t):
        [S, A, P, MI, SI, R, H] = y
        dSdt = - (rate[s][a]+rate[s][p]) * S * (Cpars[0]*A + Cpars[1]*P + Cpars[2]*MI + Cpars[3]*SI + Cpars[4]*H)/(N)
        
        dAdt = + rate[s][a] * S * (Cpars[0]*A + Cpars[1]*P + Cpars[2]*MI + Cpars[3]*SI + Cpars[4]*H)/(N) - rate[a][r] * A
        
        dPdt = + rate[s][p] * S * (Cpars[0]*A + Cpars[1]*P + Cpars[2]*MI + Cpars[3]*SI + Cpars[4]*H)/(N) - (rate[p][mi]+rate[p][si]) * P
        
        dMIdt = + rate[p][mi] * P - rate[mi][r] * MI
        
        dSIdt = + rate[p][si] * P - (rate[si][r] + rate[si][h]) * SI
        
        dHdt = + rate[si][h] * SI - rate[h][r] * H
        
        dRdt = + rate[a][r] * A + rate[mi][r] * MI + rate[h][r] * H + rate[si][r] * SI
        
#         print(dSdt+dAdt+dPdt+dMIdt+dSIdt+dHdt+dRdt)
        return dSdt, dAdt, dPdt, dMIdt, dSIdt, dRdt, dHdt
    
    #########################################
    
    print(rate)
    
    y0 = n                              # Initial conditions vector
    ret = odeint(deriv, y0, t) # Integrate the SIR equations over the time grid, t.
    
    S, A, P, MI, SI, R, H = ret.T                                   # Return the transpose of the matrix

    return t, S/N, A/N, P/N, MI/N, SI/N, R/N, H/N
```

```python

T,S,A,P,MI,SI,R,H = an_soln(rate, Cpars, n0, 200)
```

```python
R0 = gamma*k_S/k_A - (1-gamma)*k_S*((-1+delta)*k_H*k_MI*k_P - k_H*(k_MI + delta*k_P)*k_SI + (-1+delta)*k_MI*k_P*k_SI*sigma)/(k_H*k_MI*k_SI*k_P)
print(R0)
plt.title("Analytical Solution (no networks)")
plt.plot(T,A+P+MI+SI+H,'firebrick',label="Total infected")
# plt.plot(T[:50],np.log10(10/10000 * np.e**(0.06*R0*T[0:50])),'royalblue',label="Reference")
plt.plot(T[:40],0.001*R0**(T[:40]/6),label=r"$I(0)"+str(round(R0,2))+"^{t/5}$")
plt.ylabel("Fraction of population")
plt.xlabel("Time (Days)")
plt.legend()
plt.yscale('log',base=np.e)
plt.show()
```

```python
plt.plot(T, (A+P+MI+SI+H)/0.001, label="N(t)/N0")
plt.plot(T[0:30], 3**(T[:30]/8),lw=3,label="Tau_d=5")
plt.yscale("log",base=2)
# plt.xlim(-1,20)
plt.legend()
```

```python
plt.plot(R,S, label="S(t)")
plt.plot(R,S[0]*np.exp(-R0*R),'--',lw=2, label=r"$S(0) \exp(-R_0 R(t))$",color='firebrick')
plt.legend()
```
