```python
import numpy as np
import tkinter as tk
import tkinter.ttk as ttk
from ttkthemes import ThemedTk
import glob
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# s=ttk.Style()

# s.theme_use('clam')


def run():
    
    gamma = probA.get()/100
    delta = 1- (probSI.get()/100)
    sigma = probH.get()/100
    
    k_S = round(exitS.get(),3)
    k_A = round(1/exitA.get(),3)
    k_P = round(1/exitP.get(),3)
    k_MI= round(1/exitSIMI.get(),3)
    k_SI= round(1/exitSIMI.get() / (1 - sigma),3)
    k_H = round(1/exitSIMI.get(),3)
    
    
    rat_fraction = round(RATFrac.get(),3)
    
    rat_delay = round(RATDelay.get(),3)
    pcr_delay = round(PCRDelay.get(),3)
    
    lock_homes = round(QH.get(),3)
    quarantine_when_sample_taken = round(QST.get(),3)
    
    test_frac = round(testRate.get(),3)
    begin_at  = round(beginAt.get(),3)
    
    tlct = round(LocConfFor.get(),3)
    
    hospital_damping = round(Hpar.get(),3)
    quarantine_damping = round(Qpar.get(),3)
    
    fun.run.argtypes = [ctypes.c_double, ctypes.c_double, ctypes.c_double, ctypes.c_double, ctypes.c_double, ctypes.c_double, ctypes.c_double,ctypes.c_double, ctypes.c_double,
                       ctypes.c_bool, ctypes.c_bool, ctypes.c_double, ctypes.c_double,
                       ctypes.c_double, ctypes.c_double, ctypes.c_double, ctypes.c_double, ctypes.c_double, ctypes.c_double]
    
#     print(k_S, gamma, k_A, k_P, delta, k_MI, k_SI, sigma, k_H,
#             lock_homes, quarantine_when_sample_taken, rat_delay, pcr_delay,
#              rat_fraction, tlct, hospital_damping, quarantine_damping, begin_at, test_frac)
     
    for i in range(iterations.get()):
        print("Running code, Iteration", i+1)
        fun.run(k_S, gamma, k_A, k_P, delta, k_MI, k_SI, sigma, k_H,
                lock_homes, quarantine_when_sample_taken, rat_delay, pcr_delay,
                 rat_fraction, tlct, hospital_damping, quarantine_damping, begin_at, test_frac)
    
    plot()
    
    




root= ThemedTk(theme="radiance") 
root.title('7 Compartment Model -- Targetted Testing')
root.rowconfigure(0, weight=100)

content = ttk.Frame(root)

exitS = tk.DoubleVar()
exitS.set(0.25)
exitS_label = tk.Label(root, text="Exit S").grid(column=0,row=0,columnspan=2,sticky='s')
exitS_text = tk.Entry(root, textvariable=exitS, width=5, justify='center').grid(column=0,row=1,columnspan=1,sticky='s')
exitS_scale = tk.Scale(root, from_=0, to=1, orient='horizontal',variable=exitS,resolution=0.05).grid(column=1,row=1,columnspan=1,sticky='s')
# exitS_text.bind("<Return>", lambda event: scale.set(exitS.get()))

# lambdaS = tk.Label(root, text="Exit S\n"+str(exitS.get())).grid(column=2,row=1,columnspan=2,sticky='s')

probA = tk.DoubleVar()
probA.set(50)
probA_label = tk.Label(root, text="Probability of Asymptomatics").grid(column=0,row=2,columnspan=2,sticky='s',pady=(10,0))
probA_text  = tk.Entry(root, textvariable=probA, width=5, justify='center').grid(column=0,row=3,columnspan=1,sticky='s')
probA_scale = tk.Scale(root, from_=0, to=100, orient='horizontal',variable=probA,resolution=1).grid(column=1,row=3,columnspan=1,sticky='s')

# lambdaA = tk.Label(root, text="Exit A\n"+str(round(1/exitA.get(),3))).grid(column=2,row=2,columnspan=2,sticky='s')

exitA = tk.DoubleVar()
exitA.set(7)
exitA_label = tk.Label(root, text="Days for Asymptotics to Recover").grid(column=0,row=4,columnspan=2,sticky='s',pady=(10,0))
exitA_text  = tk.Entry(root, textvariable=exitA, width=5, justify='center').grid(column=0,row=5,columnspan=1,sticky='s')
exitA_scale = tk.Scale(root, from_=0.1, to=10, orient='horizontal',variable=exitA,resolution=0.1).grid(column=1,row=5,columnspan=1,sticky='s')



exitP = tk.DoubleVar()
exitP.set(2)
exitP_label = tk.Label(root, text="Days to Develop Symptoms").grid(column=0,row=6,columnspan=2,sticky='s',pady=(10,0))
exitP_text  = tk.Entry(root, textvariable=exitP, width=5, justify='center').grid(column=0,row=7,columnspan=1,sticky='s')
exitP_scale = tk.Scale(root, from_=0.1, to=10, orient='horizontal',variable=exitP,resolution=0.1).grid(column=1,row=7,columnspan=1,sticky='s')

# lambdaP = tk.Label(root, text="Exit P\n"+str(round(1/exitP.get(),3))).grid(column=2,row=3,columnspan=2,sticky='s')

exitSIMI = tk.DoubleVar()
exitSIMI.set(10)
exitSIMI_label = tk.Label(root, text="Days for Symptomatics to Recover").grid(column=0,row=8,columnspan=2,sticky='s',pady=(10,0))
exitSIMI_text  = tk.Entry(root, textvariable=exitSIMI, width=5, justify='center').grid(column=0,row=9,columnspan=1,sticky='s')
exitSIMI_scale = tk.Scale(root, from_=0.1, to=30, orient='horizontal',variable=exitSIMI,resolution=0.1).grid(column=1,row=9,columnspan=1,sticky='s')

# lambdaSIMI = tk.Label(root, text="Exit SI/MI/H->R\n",textvariable=1/exitSIMI.get()).grid(column=2,row=4,columnspan=2,sticky='s')


probSI = tk.DoubleVar()
probSI.set(15)
probSI_label = tk.Label(root, text="Probability of Severe Symptoms").grid(column=0,row=10,columnspan=2,sticky='s',pady=(10,0))
probSI_text  = tk.Entry(root, textvariable=probSI, width=5, justify='center').grid(column=0,row=11,columnspan=1,sticky='s')
probSI_scale = tk.Scale(root, from_=0, to=100, orient='horizontal',variable=probSI,resolution=1).grid(column=1,row=11,columnspan=1,sticky='s')


probH = tk.DoubleVar()
probH.set(80)
probH_label = tk.Label(root, text="Probability this leads to Hospitalisation").grid(column=0,row=12,columnspan=2,sticky='s',pady=(10,0))
probH_text  = tk.Entry(root, textvariable=probH, width=5, justify='center').grid(column=0,row=13,columnspan=1,sticky='s')
probH_scale = tk.Scale(root, from_=0, to=100, orient='horizontal',variable=probH,resolution=1).grid(column=1,row=13,columnspan=1,sticky='s')




Hpar = tk.DoubleVar()
Hpar.set(0.1)
Hpar_label = tk.Label(root, text="Hospital Damping").grid(column=0,row=14,columnspan=2,sticky='s',pady=(10,0))
Hpar_text  = tk.Entry(root, textvariable=Hpar, width=5, justify='center').grid(column=0,row=15,columnspan=1,sticky='s')
Hpar_scale = tk.Scale(root, from_=0, to=1, orient='horizontal',variable=Hpar,resolution=0.05).grid(column=1,row=15,columnspan=1,sticky='s')

Qpar = tk.DoubleVar()
Qpar.set(0.1)
Qpar_label = tk.Label(root, text="Quarantine Damping").grid(column=0,row=16,columnspan=2,sticky='s',pady=(10,0))
Qpar_text  = tk.Entry(root, textvariable=Qpar, width=5, justify='center').grid(column=0,row=17,columnspan=1,sticky='s')
Q= tk.Scale(root, from_=0, to=1, orient='horizontal',variable=Qpar,resolution=0.05).grid(column=1,row=17,columnspan=1,sticky='s')



Testing_label = tk.Label(root, text="TESTING PARAMETERS").grid(column=3,row=13,columnspan=6,sticky='s',pady=(10,0))


testRate = tk.DoubleVar()
testRate.set(10)
testRate_label = tk.Label(root, text="Daily Testing Rate").grid(column=3,row=14,columnspan=2,sticky='s',pady=(10,0))
testRate_text  = tk.Entry(root, textvariable=testRate, width=5, justify='center').grid(column=3,row=15,columnspan=1,sticky='s')
testRate_scale = tk.Scale(root, from_=0, to=100, orient='horizontal',variable=testRate,resolution=0.05).grid(column=4,row=15,columnspan=1,sticky='s')

beginAt = tk.DoubleVar()
beginAt.set(10)
beginAt_label = tk.Label(root, text="Percentage Infected\nbefore Testing").grid(column=3,row=16,columnspan=2,sticky='s',pady=(10,0))
beginAt_text  = tk.Entry(root, textvariable=beginAt, width=5, justify='center').grid(column=3,row=17,columnspan=1,sticky='s')
beginAt_scale = tk.Scale(root, from_=0, to=100, orient='horizontal',variable=beginAt,resolution=0.05).grid(column=4,row=17,columnspan=1,sticky='s')


RATFrac = tk.DoubleVar()
RATFrac.set(0)
RATFrac_label = tk.Label(root, text="Fraction of RAT").grid(column=5,row=14,columnspan=2,sticky='s',pady=(10,0))
RATFrac_text  = tk.Entry(root, textvariable=RATFrac, width=5, justify='center').grid(column=5,row=15,columnspan=1,sticky='s')
RATFrac_scale = tk.Scale(root, from_=0, to=100, orient='horizontal',variable=RATFrac,resolution=0.05).grid(column=6,row=15,columnspan=1,sticky='s')

RATDelay = tk.DoubleVar()
RATDelay.set(1)
RATDelay_label = tk.Label(root, text="RAT Delay").grid(column=5,row=16,columnspan=2,sticky='s',pady=(10,0))
RATDelay_text  = tk.Entry(root, textvariable=RATDelay, width=5, justify='center').grid(column=6,row=16,columnspan=1,sticky='s')

PCRDelay = tk.DoubleVar()
PCRDelay.set(1)
PCRDelay_label = tk.Label(root, text="PCR Delay").grid(column=5,row=17,columnspan=2,sticky='s',pady=(10,0))
PCRDelay_text  = tk.Entry(root, textvariable=PCRDelay, width=5, justify='center').grid(column=6,row=17,columnspan=1,sticky='s')



Quarantine_label = tk.Label(root, text="QUARANTINE\nPARAMETERS").grid(column=7,row=13,columnspan=3,sticky='s',pady=(10,0))

QH = tk.BooleanVar()
QH.set(0)
QH_check = tk.Checkbutton(root, text='Quarantine Household',variable=QH, onvalue=1, offvalue=0).grid(column=7,row=15,columnspan=3,sticky='s',pady=(10,0))

QST = tk.BooleanVar()
QST.set(0)
QST_check = tk.Checkbutton(root, text='Quarantine when Sample Taken',variable=QST, onvalue=1, offvalue=0).grid(column=7,row=16,columnspan=3,sticky='s',pady=(10,0))

LocConfFor = tk.DoubleVar()
LocConfFor.set(14)
LocConfFor_label = tk.Label(root, text="Location\nConfined\nFor").grid(column=7,row=17,columnspan=1,sticky='s',pady=(10,0))
LocConfFor_text  = tk.Entry(root, textvariable=LocConfFor, width=5, justify='center').grid(column=8,row=17,columnspan=1,sticky='s')


iterations = tk.IntVar()
iterations.set(2)
tk.Label(root, text="Iterations").grid(column=0,row=18,columnspan=3,sticky='s',pady=(10,0))
tk.Entry(root, textvariable=iterations, width=5, justify='center').grid(column=1,row=18,columnspan=3,sticky='s')


tk.Button(root, text='RUN!',bg='royalblue', fg='navy', width=10,command=run).grid(column=0,row=19,columnspan=1,pady=10)
tk.Button(root, text='PLOT',bg='lightseagreen', fg='darkgreen', width=10,command=plot).grid(column=1,row=19,columnspan=1,pady=10)
tk.Button(root, text='KILL!',bg='firebrick', fg='orange', width=10,command=root.destroy).grid(column=2,row=19,columnspan=1,pady=10)

content.grid(column=2, row=2,columnspan=10,rowspan=10)

def plot():
    print("Plotting")
    files = glob.glob("./Targetted*.txt")
    
    figure1 = plt.Figure(figsize=(10,5), dpi=100)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, content)
    bar1.get_tk_widget().grid(column=0,row=0,columnspan=20)

    avg = np.zeros((201,11),dtype=float)
    counter = 0
    
    for i in range(0,len(files)):
        file = np.genfromtxt(files[i])
        
        t = file[:,0]
        S = file[:,1]
        I = file[:,2]+file[:,3]+file[:,4]+file[:,5]+file[:,7]
        R = file[:,6]

        ax1.plot(t,S,'--',color='navy',alpha=0.2)
        ax1.plot(t,I,'--',color='firebrick',alpha=0.2)
        ax1.plot(t,R,'--',color='darkgreen',alpha=0.2)
        
        if(R[-1]>100):
            avg += file
            counter += 1
    
    avg = avg/counter
    
    avg_t = avg[:,0]
    avg_S = avg[:,1]
    avg_I = avg[:,2]+file[:,3]+file[:,4]+file[:,5]+file[:,7]
    avg_R = avg[:,6]
    
    ax1.plot(t,avg_S,color='navy',alpha=1, label="Susceptible "+str(round(avg_S[-1])))
    ax1.plot(t,avg_I,color='firebrick',alpha=1)
    
    I_peak = max(avg_I)
    t_peak = avg_t[np.where(avg_I==I_peak)]
    
    ax1.text(2+t_peak,2+I_peak, "Peak: "+str(round(I_peak,2)),color='firebrick')
    ax1.plot(t,avg_R,color='darkgreen',alpha=1, label="Recovered "+str(round(avg_R[-1])))
    
    ax1.set_xlabel("Time (Days)")
    ax1.set_ylabel("Number of individuals")
    ax1.legend()






root.mainloop()
```
