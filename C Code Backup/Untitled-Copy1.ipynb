```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
S = 0
A = 1
P = 2
MI= 3
SI= 4
R = 5
H = 6
```

```python
def plot_all(s):
    
    files = glob.glob("./Targetted-Testing_BeginAt_"+s+"*.txt")
    
    avgs = np.zeros((201,11))
    rec = []
    
    for i in range(0,len(files)):
#         print(files[i])
        f = np.genfromtxt(files[i])
#         plt.plot(f[:,0],f[:,2]+f[:,3]+f[:,4]+f[:,5]+f[:,7], '--',color='darkgreen',alpha=0.3)
#         plt.plot(f[:,0],f[:,6])
    
        avgs += f
        rec.append(f[-1][6])
    print(len(files))
    avgs = avgs/len(files)
#     plt.plot(avgs[:,0],avgs[:,2]+avgs[:,3]+avgs[:,4]+avgs[:,5]+avgs[:,7], color='darkgreen')
    plt.plot(avgs[:,0],avgs[:,6], color='lime',label="Average "+str(np.average(rec)))
```

```python
def plot_ran():
    
    files = glob.glob("./Targetted*.txt")
    
    avgs = np.zeros((201,11))
    rec = []
    
    f = np.genfromtxt(np.random.choice(files))
    plt.plot(f[:,0], f[:,A+1]+f[:,P+1]+f[:,MI+1]+f[:,SI+1]+f[:,H+1], color='firebrick',label="Peak Infected "+str(max(f[:,2]+f[:,3]+f[:,4]+f[:,5]+f[:,7])),alpha=0.2)
    plt.plot(f[:,0],f[:,R+1], '--',color='darkgreen',label= "Recovered "+str(f[-1][6]),alpha=0.2)
    plt.plot(f[:,0],f[:,S+1], '--',color='navy',label= "Susceptible "+str(f[-1][1]),alpha=0.2)
#     plt.legend()
    plt.xlim(0,200)
```

```python
def plot_three(s):
    
    files = glob.glob("./"+s+"/Targetted*.txt")
    
    avgs = np.zeros((201,11))
    rec = []
    counter = 0
    for i in range(len(files)):
    
        f = np.genfromtxt(files[i])
        if(f[-1][R+1]>100):
            counter +=1
            plt.plot(f[:,0], f[:,A+1]+f[:,P+1]+f[:,MI+1]+f[:,SI+1]+f[:,H+1],'--', color='firebrick',alpha=0.2)
            plt.plot(f[:,0],f[:,R+1], '--',color='darkgreen',alpha=0.2)
            plt.plot(f[:,0],f[:,S+1], '--',color='navy',alpha=0.2)

            avgs += f
#     plt.legend()
    avgs = avgs/counter
    
    plt.plot(avgs[:,0],avgs[:,A+1]+avgs[:,P+1]+avgs[:,MI+1]+avgs[:,SI+1]+avgs[:,H+1], color='firebrick',label="Peak Infected "+str(max(avgs[:,2]+avgs[:,3]+avgs[:,4]+avgs[:,5]+avgs[:,7])))
    plt.plot(avgs[:,0],avgs[:,R+1],color='darkgreen',label= "Recovered "+str(avgs[-1][6]),alpha=1)
    plt.plot(avgs[:,0],avgs[:,S+1],color='navy',label= "Susceptible "+str(avgs[-1][1]),alpha=1)
    
    
    plt.xlim(0,200)
```

```python
def plot_three_SK(s):
    S = 0
    A = 1
    P = 2
    MI= 3
    SI= 4
    R = 6
    H = 5
    
    
    
    files = glob.glob("./"+s+"/SK*.txt")
#     print(files)
    avgs = np.zeros((200,8))
    rec = []
    
    for i in range(len(files)):
        f = np.genfromtxt(files[i])
#         print(len(f))
        plt.plot(f[:,0], f[:,A+1]+f[:,P+1]+f[:,MI+1]+f[:,SI+1]+f[:,H+1],'--', color='firebrick',alpha=0.2)
        plt.plot(f[:,0],f[:,R+1], '--',color='darkgreen',alpha=0.2)
        plt.plot(f[:,0],f[:,S+1], '--',color='navy',alpha=0.2)
        
        avgs += f
#     plt.legend()
    avgs = avgs/len(files)
    
    plt.plot(avgs[:,0],avgs[:,A+1]+avgs[:,P+1]+avgs[:,MI+1]+avgs[:,SI+1]+avgs[:,H+1], color='firebrick',label="Peak Infected "+str(max(avgs[:,A+1]+avgs[:,P+1]+avgs[:,MI+1]+avgs[:,SI+1]+avgs[:,H+1])))
    plt.plot(avgs[:,0],avgs[:,R+1],color='darkgreen',label= "Recovered "+str(avgs[-1][R+1]),alpha=1)
    plt.plot(avgs[:,0],avgs[:,S+1],color='navy',label= "Susceptible "+str(avgs[-1][S+1]),alpha=1)
    
    
    plt.xlim(0,200)
```
